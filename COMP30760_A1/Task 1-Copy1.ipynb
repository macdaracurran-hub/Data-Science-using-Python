 
# Using the API key to fetch Now-Playing Movies data

```python
key = 'c5aa1952990374cd0dd1342324866ccd'
link = f"https://api.themoviedb.org/3/movie/now_playing?api_key={key}&language=en-US&page=1"
response = requests.get(link)
response
```

There is a response at 200, this means everything is working properly. The data is in the key results, so I will print the results key.<br>


```python
response.json()["results"]
```

```python
data = pd.DataFrame(response.json()["results"])[['id','title','adult','popularity','release_date','vote_average','vote_count']]
data.shape
```

The data of all 97 pages is needed, so a for loop is used to request different pages from the website repeatedly.


```python
df = data.copy()
for i in range(1, 98):
    response = requests.get("https://api.themoviedb.org/3/movie/now_playing?api_key=c5aa1952990374cd0dd1342324866ccd&language=en-US&page={}".format(i))
    print(response, ' at page {}'.format(i))
    temp_df = pd.DataFrame(response.json()["results"])[['id','title','adult','popularity','release_date','vote_average','vote_count']]
    df = pd.concat([df, temp_df], axis=0, ignore_index=True)
```
